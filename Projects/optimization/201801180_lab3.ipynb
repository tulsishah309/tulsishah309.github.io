```python
import numpy as np
import math

ans=[];   
    
row,column=(int(row) for row in input().split(" "));  

matrix=[];

for i in range (row):
    x= [int(x) for x in input().split(" ")]   
    matrix.append(x);

a=[];
b=[];


for i in range(row):
    d=[];
    for j in range(column):
        d.append(0);
    ans.append(d);
        
        

a= [int(a) for a in input().split(" ")]    
b= [int(b) for b in input().split(" ")]    

checkrow=[];    
checkcol=[];    
for i in range(row):
    checkrow.append(0);
    
for i in range(column):
    checkcol.append(0);

loop=0;



while(loop==0):
    
    rowMin=[];    
    colMin=[];   
    
    for i in range(row):
        rowMin.append(0);
    
    for i in range(column):
        colMin.append(0);
    
   

    flag=0;
    for i in range(row):
        flag=-1;
        Min1=math.inf;     
        Min2=math.inf;
        for j in range(column):
            if(checkcol[j]==0):
                if(int(list(matrix)[i][j])<Min1):
                    Min1=matrix[i][j];
                    flag=j;
        
        for j in range(column):
            if(checkcol[j]==0):
                if(int(list(matrix)[i][j])<Min2 and j!=flag):
                    Min2=matrix[i][j];
                    
        rowMin[i]=Min2-Min1;
        
    for i in range(column):
        flag=-1;
        Min1=math.inf;
        Min2=math.inf;
        for j in range(row):
            if(checkcol[j]==0):
                if(int(list(matrix)[j][i])<Min1):
                    Min1=matrix[j][i];
                    flag=j;
        
        for j in range(row):
            if(checkcol[j]==0):
                if(int(list(matrix)[j][i])<Min2 and j!=flag):
                    Min2=matrix[j][i];
                    
        colMin[i]=Min2-Min1;
        
    Max=0;
    r=0;
    c=0;
    
    index1=0;
    for i in range(row):
        if(checkrow[i]==0):
            Max=max(Max,rowMin[i]);
            if(Max==rowMin[i]):
                index1=i;
                r=1;
            
    for i in range(column):
        if(checkcol[i]==0):
            Max=max(Max,colMin[i]);
            if(Max==colMin[i]):
                index1=i;
                r=0;c=1;
                
    index2=-1;          
    if(r==1):
        Min=math.inf;
        for j in range(column):
            if(Min>matrix[index1][j] and checkcol[j]==0):
                Min=matrix[index1][j];
                index2=j;
                
    if(c==1):
        Min=math.inf;
        for j in range(row):
            if(Min>matrix[j][index1] and checkrow[j]==0):
                Min=matrix[j][index1];
                index2=j;
                    
    if(r==1):           
        if(a[index1]>b[index2]):
            ans[index1][index2]=b[index2];
            a[index1]=a[index1]-b[index2];
            b[index2]=0;
            checkcol[index2]=1;

        else:
            ans[index1][index2]=a[index1];
            b[index2]=b[index2]-a[index1];
            a[index1]=0;
            checkrow[index1]=1;
    else :
        index1,index2=index2,index1;
        #print(index1,' ',index2) 
        if(a[index1]>b[index2]):
            ans[index1][index2]=b[index2];
            a[index1]=a[index1]-b[index2];
            b[index2]=0;
            checkcol[index2]=1;

        else:
            ans[index1][index2]=a[index1];
            b[index2]=b[index2]-a[index1];
            a[index1]=0;
            checkrow[index1]=1;
        
     
    cnt1=0;cnt2=0;
    for i in range(row):
        if(checkrow[i]==0):
            cnt1=cnt1+1;
            break;
            
    for i in range(column):
        if(checkcol[i]==0):
            cnt2=cnt2+1;
            break;
            
    if((cnt1==1 and cnt2==0) or (cnt1==0 and cnt2==1)):
        loop=1;
        
   

bf=0;

for i in range(row):
    for j in range(column):
        bf=bf+(ans[i][j]*matrix[i][j]);
        
print("Initial bfs");

for i in range(row):
    for j in range(column):
        if(ans[i][j]!=0):
            x='x'+str(i+1)+str(j+1)
            print(x,'=',ans[i][j]);
print('Cost=',bf);


MODI=0;
while(MODI==0):
    c=[];

    for i in range(row):
        d=[];
        for j in range(column):
            d.append(0);
        c.append(d);   

    for i in range(row):
        for j in range(column):
            if(ans[i][j]!=0):
                c[i][j]=matrix[i][j];

    MaxCol=0;MaxRow=0;rowIndex=0;colIndex=0;

    for i in range(row):
        cnt=0;
        for j in range(column):
            if(ans[i][j]!=0):
                cnt=cnt+1;
        MaxRow=max(MaxRow,cnt);
        if(MaxRow==cnt):
            rowIndex=i;

    for i in range(column):
        cnt=0;
        for j in range(row):
            if(ans[j][i]!=0):
                cnt=cnt+1;
        MaxCol=max(MaxCol,cnt);
        if(MaxCol==cnt):
            colIndex=i;

    maxIndex=-1;   
    u=[];  #Ui's
    v=[];  #Vi's

    for i in range(row):
        u.append(-1);

    for i in range(column):
        v.append(-1);

    opRow=[];
    opCol=[];
    for i in range(row):
        opRow.append(-1);

    for i in range(column):
        opCol.append(-1);

    #print(MaxCol,' ',MaxRow)
    if(MaxRow>MaxCol):
        maxIndex=rowIndex;
        u[maxIndex]=0;
        opRow[maxIndex]=1;          
    else:
        maxIndex=colIndex;
        v[maxIndex]=0;
        opCol[maxIndex]=1;

    count=1;

    if(maxIndex==MaxRow):
        for i in range(column):
            if(ans[maxIndex][i]!=0):
                v[i]=matrix[maxIndex][i];
                count=count+1;
                opCol[i]=1;


    if(maxIndex==MaxCol):
        for i in range(row):
            if(ans[i][maxIndex]!=0):
                u[i]=matrix[i][maxIndex];
                count=count+1;
                opRow[i]=1;



    while(count!=row+column):
        for i in range(row):
            if(opRow[i]==-1):
                for j in range(column):
                    if(opCol[j]==1 and ans[i][j]!=0):
                        u[i]=matrix[i][j]-v[j];
                        count=count+1;
                        opRow[i]=1;
                        break;

        for j in range(column):
            if(opCol[j]==-1):
                for i in range(row):
                    if(opRow[i]==1 and ans[i][j]!=0):
                        v[j]=matrix[i][j]-u[i];
                        count=count+1;
                        opCol[j]=1;
                        break;
   
    delta=[];      #delta matrix to see negative value
    for i in range(row):
        d=[];
        for j in range(column):
            d.append(0);
        delta.append(d);

    for i in range(row):
        for j in range(column):
            if(ans[i][j]==0):
                delta[i][j]=matrix[i][j]-(u[i]+v[j]);  #delta

    flag=1;


    r_n=0;
    c_n=0;
    for i in range(row):
        for j in range(column):
            if(delta[i][j]<0):
                flag=0;
                r_n=i;
                c_n=j;
                break;
        if(flag==0):
            break;

   
    row_c={};
    col_r={};
    for i in range(row):
        for j in range(column):
            if(ans[i][j]!=0):
                if i in row_c:
                    row_c[i].append(j);
                else:
                    row_c[i]=[j];
                if j in col_r:
                    col_r[j].append(i);
                else :
                    col_r[j]=[i];   
    
    if(flag==1):
        print('Optimal solution');
        Opcost=0;
        for i in range(row):
            for j in range(column):
                if(ans[i][j]!=0):
                    s='x'+str(i+1)+str(j+1);
                    print(s,'=',ans[i][j]);
                    Opcost=Opcost+ans[i][j]*matrix[i][j];
        print('Optimal cost=',Opcost);
        MODI=1;
        break;
    else:
        loop_p=[];    #to find loop in matrix
        for i in row_c[r_n]:
            t1=[[r_n,c_n],[r_n,i]];
            loop_p.append(t1);
        for i in col_r[c_n]:
            t1=[[r_n,c_n],[i,c_n]];
            loop_p.append(t1);
        row_c[r_n].append(c_n);
        col_r[c_n].append(r_n);

        f=0;
        cnt=0;

        while(1):
            loop_array=np.array(loop_p);
            for i in range(0,len(loop_array)):
                t1=loop_array[i].tolist();
                t2=np.array(t1);
                last=t1[-1];
                loop_p.remove(t1);

                if(cnt%2==0):
                    for j in col_r[last[1]]:
                        if((len(t1)>=4) and ([j,last[1]]==[r_n,c_n])):
                            f=1;
                            break;
                        if([j,last[1]] in t1):
                            continue;
                        else:
                            j2 = [j3 for j3 in t1 if j3[1] == last[1]]
                            if(len(j2)<2):
                                t1.append([j,last[1]])
                                loop_p.append(t1);
                                t1=t2.tolist();
                else:
                    for j in row_c[last[0]]:
                        if((len(t1)>=4) and ([last[0],j]==[r_n,c_n])):
                            f=1;
                            break;
                        if([last[0],j] in t1):
                            continue;
                        else:
                            j2 = [j3 for j3 in t1 if j3[0] == last[0]]
                            if(len(j2)<2):
                                t1.append([last[0],j])
                                loop_p.append(t1);

                                t1=t2.tolist();
                if(f):
                    t1.append([r_n,c_n]);
                    loop_p.append(t1);
                    break;


            if(f):
                break;
            cnt=cnt+1;

    Min=math.inf;
   
    i=1;
    loop_array=np.array(loop_p);
    t1=loop_array[0].tolist();
    while(i<len(t1)-1):
        t2=np.array(t1);
        c=t2[i];
        Min=min(Min,ans[c[0]][c[1]]);
        
        i=i+2;

    i=0;
    while(i<len(t1)-1):
        t2=np.array(t1);
        c=t2[i];
        if(i%2!=0):
            ans[c[0]][c[1]]=ans[c[0]][c[1]]-Min;
        else:
            ans[c[0]][c[1]]=ans[c[0]][c[1]]+Min;
        i=i+1;

    

```
